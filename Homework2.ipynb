```python
import pandas as pd

df1 = pd.read_csv("/content/Dataset Generation (2024) (Responses) - Form Responses 1.csv")
df2 = pd.read_csv("/content/Dataset Generation (Fardina) (Responses) - Form Responses 1.csv")
df3 = pd.read_csv("/content/Dataset Generation (Max) (Responses) - Form Responses 1.csv")
df4 = pd.read_csv("/content/Dataset Generation (Spring 2025) (Responses) - Form Responses 1.csv")
```

```python
df1.head(10)
df2.head(10)
df3.head(10)
df4.head(10)
```

cleaning the data by filling in the missing values and droping the dublicates

```python
datasets = [df1, df2, df3, df4]
combined_data = pd.concat(datasets, ignore_index=True)


combined_data.drop_duplicates(inplace=True)
```

```python
combined_data.head(500)
```

```python
combined_data.columns
```

```python
combined_data['How old are you?'] = pd.to_numeric(combined_data['How old are you?'], errors='coerce')

median_age = combined_data['How old are you?'].median()
combined_data['How old are you?'].fillna(median_age, inplace=True)
```

```python
combined_data['How old are you?'].value_counts()
```

```python
combined_data['What year are you?'] = combined_data['What year are you?'].fillna('Unknown')
combined_data['What year are you?'].value_counts()
```

```python
combined_data['You could describe the adults you grew up with as...'] = combined_data['You could describe the adults you grew up with as...'].fillna('Unknown')
combined_data['You could describe the adults you grew up with as...'].value_counts()
```

```python
combined_data['What bests represents your gender?'] = combined_data['What bests represents your gender?'].replace("Famale", "Female")

combined_data['What bests represents your gender?'] = combined_data['What bests represents your gender?'].replace({
    "Male": "Male",
    "Female": "Female",
    "Non-binary / other": "Non-binary",
    "M": "Male",
    "F": "Female",
})

combined_data['What bests represents your gender?'] = combined_data['What bests represents your gender?'].fillna('Unknown')

print(combined_data['What bests represents your gender?'].value_counts())
```

```python


combined_data['You could describe yourself as...'] = combined_data['You could describe yourself as...'].replace({
    "Don't know / It's complicated": "Neutral",
    "Mildly liberal": "Liberal",
    "Mildly conservative": "Conservative",
    "Strongly liberal": "Liberal",
    "Strongly conservative": "Conservative"
})
combined_data['You could describe yourself as...'] = combined_data['You could describe yourself as...'].fillna('Unknown')
combined_data['You could describe yourself as...'].value_counts()
```

```python
combined_data['How would you rate your religiousness?'] = combined_data['How would you rate your religiousness?'].replace({
    "Somewhat spiritual": "Somewhat religious",
    "Not spiritual at all": "Not religious at all",
    "Strongly spiritual": "Strongly religious",
    "Somewhat religious": "Somewhat religious",
    "Strongly religious": "Strongly religious",
})

combined_data['How would you rate your religiousness?'] = combined_data['How would you rate your religiousness?'].fillna('Unknown')

combined_data['How would you rate your religiousness?'].value_counts()
```

```python
scenario_columns = [
    "My boyfriend is a doctor. Lately he's been complaining about pain in his right knee and constantly taking TONS of ibuprofen to treat (like 15-20 pills a day). However, it's only gotten worse. I've told him he needs to see someone about it but he told me it was fine. So I told him that if this is how he treats herself, I'd be concerned about being his patient. He was shocked and told me that was uncalled for. Am I a jerk?",
    "My son is getting married soon. Due to a complicated past, we've only just met recently. We've been building a relationship the last few years though and I am lucky to have him in my life now. Recently, he asked me to do a reading with his father while I am honored I told him no. His step-mother practically raised him from a baby and I think that she deserves it far more than me. He was really upset and hurt when I said no. Am I a jerk?",
    "I'm a trust fund kid; I get a healthy 'allowance' from my parents, but I mostly sock it away since I don't really feel like I deserve it. I live with my boyfriend, who comes from a poor family, and we split things 50-50. Recently, he found out how much money my parents send me each month and freaked out, saying it wasn't fair that I was making him split things 50/50. I told him it was my money, and it only seems fair if we cover half the expenses. Am I a jerk?",
    "My husband and I have separate finances, but I pay for almost everything. My son starts school next year, and I'm planning on sending him to a private school. My husband's son currently goes to a much less nice public school. My husband says I'm a jerk for not paying for his son to go to private school as well.",
    "I saw a poster for a lost cat advertising a 500 dollar reward. I saw the cat, tracked it down, and called the owner. When I met with them, they said there wasn't actually a reward, they had just listed one because they thought it would motivate people to find the cat. I kept the cat until they coughed up the reward money. Am I a jerk?",
    "My brother's nine year old son is poorly behaved. One day, my brother dropped the son off on my doorstep without my consent and drove off, texting me that he needed me to watch his son for the day. The son complained of being hungry, so I gave him the leftovers I had. He said he didn't like them, but I told him that's all there was. He put them in the microwave until they burned, then told me he couldn't eat them and we would have to go out and get food. I told him that was what there was to eat, and if he ruined it he'd have to wait for his dad to get back. Later, everyone yelled at me for 'starving' a child. Am I a jerk?",
    combined_data.columns[combined_data.columns.str.startswith("My parents want us to come out for their anniversary, and bought my brother and I tickets on the same flight.")][0],
    "I'm a single dad with four kids, one of whom has a different mother from the other three. I get a lot of child support for one of them, and none for the other 3. 2/3rds of the child support for the eldest go towards his expenses and college fund. The other third goes towards my other children. My friend recently found out and told me it was inappropriate to use the child support intended for the one on the other three. Am I a jerk?",
    "I have a child with a father who never wanted anything to do with them. I make enough money to cover my expenses, but I constantly check up on the child's father to make sure he's paying adequate child support. Periodically I either hire an investigator or her dad just tells me he got a raise and I go to court to adjust the child support accordingly. The last time I did this, he didn't show up for court and I apparently overestimated how much his raise was and I got a default judgment. He called me up and practically pleaded with me to lower his payments after I said no he insulted me and my son. Am I a jerk?",
    "One of my children wants to go to an expensive school to become a dentist. I told them I'd be fine paying for it. The other one wants to go to a similarly expensive school to become a teacher. I told her that I wouldn't pay for him schooling unless he picked a career path that will make at least 100,000 a year. Am I a jerk?",
    "I was in a conflict with my father-in-law's girlfriend, in which I made a snide comment about she's never paid child support to his ex. Later, she called me and told me if I ever said anything like that ever again she'd sue me for defamation. I took the threat seriously, as she is a lawyer, and uninvited him for our wedding and told my family I couldn't be around him anymore. Am I a jerk?",
    "Some of my relatives refuse to come to my wedding, since they don't approve of our 'lifestyle'. I would like to donate the money I will save to an LBGTQ organization in their name. Im hoping that I or the organization will be able to send them a receipt/thank you for the donation. Would I be a jerk?",
    "My wife has decided that since she cant drink because she is pregnant that I cant either. I planned to take my annual leave from work to take us (and pay for) a trip abroad. I asked my wife would the drinking rule still be in effect and she said of course. Ive now decided I wont be paying for us to take this trip. Am I a jerk?",
    "My brother is going to be a groomsman at my wedding. His hair was dyed, but he recently decided to grow it out, so parts of it are bleached and part are not. I told him it wouldn't look good in the pictures, and offered to pay for him to dye it. He said no, and I said he couldn't be in the wedding party. Am I a jerk?"
]


for col in scenario_columns:
    combined_data[col] = combined_data[col].replace({
        "Not a jerk": "Not a jerk",
        "Mildly a jerk": "Mildly a jerk",
        "Strongly a jerk": "Strongly a jerk"
    })
combined_data[scenario_columns] = combined_data[scenario_columns].fillna('Unknown')

```

```python
for col in scenario_columns:
    print(f"Value counts for '{col}':")
    print(combined_data[col].value_counts())
    print("\n" + "-"*50 + "\n")
```

```python
combined_data['Would you describe yourself as compassionate?'] = combined_data['Would you describe yourself as compassionate?'].replace({
    "yes": "Yes",
    "Yes": "Yes",
    "y": "Yes",
    "no": "No",
    "No": "No",
    "n": "No"
})

combined_data['Would you describe yourself as compassionate?'] = combined_data['Would you describe yourself as compassionate?'].fillna('Unknown')

print(combined_data['Would you describe yourself as compassionate?'].value_counts())
```

```python
import scipy.stats as stats

for col in scenario_columns:
    contingency_table_gender = pd.crosstab(combined_data['What bests represents your gender?'], combined_data[col])

    chi2_gender, p_gender, dof_gender, expected_gender = stats.chi2_contingency(contingency_table_gender)

    print(f"Chi-Square Test result for Gender and '{col}':")
    print(f"Chi2: {chi2_gender}, p-value: {p_gender}, degrees of freedom: {dof_gender}\n")


```

```python
for col in scenario_columns:
    contingency_table_religiousness = pd.crosstab(combined_data['How would you rate your religiousness?'], combined_data[col])

    chi2_religiousness, p_religiousness, dof_religiousness, expected_religiousness = stats.chi2_contingency(contingency_table_religiousness)

    print(f"Chi-Square Test result for Religiousness and '{col}':")
    print(f"Chi2: {chi2_religiousness}, p-value: {p_religiousness}, degrees of freedom: {dof_religiousness}\n")
```

```python
import matplotlib.pyplot as plt

for col in scenario_columns:
    contingency_table_gender = pd.crosstab(combined_data['What bests represents your gender?'], combined_data[col])

    contingency_table_normalized_gender = contingency_table_gender.div(contingency_table_gender.sum(axis=1), axis=0)

    contingency_table_normalized_gender.plot(kind='bar', stacked=True, figsize=(10,6))
    plt.title(f'Gender and Judgment: Responses to "{col}"')
    plt.xlabel('Gender')
    plt.ylabel('Proportion')
    plt.legend(title="Judgment")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
for col in scenario_columns:
    contingency_table_political_leaning = pd.crosstab(combined_data['You could describe yourself as...'], combined_data[col])

    chi2_political, p_political, dof_political, expected_political = stats.chi2_contingency(contingency_table_political_leaning)

    print(f"Chi-Square Test result for Political Leaning and '{col}':")
    print(f"Chi2: {chi2_political}, p-value: {p_political}, degrees of freedom: {dof_political}\n")
```

```python
for col in scenario_columns:
    scenario_distribution = combined_data[col].value_counts()

    if scenario_distribution.get("Not a jerk", 0) > scenario_distribution.get("Mildly a jerk", 0) and scenario_distribution.get("Not a jerk", 0) > scenario_distribution.get("Strongly a jerk", 0):
        print(f"Scenario '{col}' is more likely to be judged as 'Not a jerk'.")
    elif scenario_distribution.get("Strongly a jerk", 0) > scenario_distribution.get("Mildly a jerk", 0) and scenario_distribution.get("Strongly a jerk", 0) > scenario_distribution.get("Not a jerk", 0):
        print(f"Scenario '{col}' is more likely to be judged as 'Strongly a jerk'.")
    else:
        print(f"Scenario '{col}' has a more even distribution of responses.")
    print(scenario_distribution)
    print("\n" + "-"*50 + "\n")

```

```python
# Loop through each scenario and create a stacked bar chart to visualize the polarizing scenarios
for col in scenario_columns:
    # Create a contingency table for each scenario
    contingency_table_scenario = pd.crosstab(combined_data['What bests represents your gender?'], combined_data[col])

    # Normalize the contingency table for proportions
    contingency_table_normalized_scenario = contingency_table_scenario.div(contingency_table_scenario.sum(axis=1), axis=0)

    # Plotting
    contingency_table_normalized_scenario.plot(kind='bar', stacked=True, figsize=(10,6))
    plt.title(f'Polarizing Judgment: Responses to "{col}"')
    plt.xlabel('Gender')
    plt.ylabel('Proportion')
    plt.legend(title="Judgment")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```
